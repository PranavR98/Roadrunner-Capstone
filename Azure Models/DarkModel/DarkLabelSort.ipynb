## DEFINE COPY IMAGES FUNCTION
**In this version, we pull from the 'quality' column instead of the 'labels' column**

```python
def copy_images_to_label_folders(dataframe, source_directory, target_directory):
    
    dataframe = dataframe.dropna(subset=['quality'])
    
    for label in dataframe['quality'].unique():
        label_folder = os.path.join(target_directory, str(label))
        os.makedirs(label_folder, exist_ok=True)

    for index, row in dataframe.iterrows():
        image_filename = row['ext']
        label = row['quality']
        source_path = os.path.join(source_directory, image_filename)
        target_path = os.path.join(target_directory, str(label), image_filename)
        shutil.copy(source_path, target_path)
```

## Darken additional photos to make the model more robust

```python
source_dir = "Photos Upload\\pass_all"
destination_dir = "Photos Upload\\fail_all"
```

```python
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)
```

```python
files = os.listdir(source_dir)
files.sort()
```

```python
for i, filename in enumerate(files[5000:5201], start=5000):
    # Build the full file paths
    source_path = os.path.join(source_dir, filename)
    destination_path = os.path.join(destination_dir, filename)

    # Open the image
    image = Image.open(source_path)

    # Define enhancement factor based on the file range
    if 5000 <= i <= 5050:
        enhance_factor = 0.05
    elif 5051 <= i <= 5100:
        enhance_factor = 0.10
    elif 5101 <= i <= 5150:
        enhance_factor = 0.15
    elif 5151 <=i <= 5200:
        enhance_factor = 0.20
    else:
        enhance_factor = 1.0  # Default to no enhancement

    enh = ImageEnhance.Brightness(image)
    image1 = enh.enhance(enhance_factor)

    # Save the enhanced image to the destination directory
    image1.save(destination_path)
```
