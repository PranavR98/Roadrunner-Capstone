```python
def check_blurry(directory, file_num):
    file_num = str(file_num)

    files = []
    variances = []
    ff_means = []

    # Loop through each image in the directory
    for file_name in os.listdir(directory + file_num):
        # Check if the file is an image
        if file_name.endswith(".jpg") or file_name.endswith(".png"):
            files.append(file_name)

            # Load the image and convert it to grayscale
            image = cv2.imread(os.path.join(directory, file_num, file_name))
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

            # Calculate the variance of Laplacian
            laplacian = cv2.Laplacian(gray, cv2.CV_64F)
            variance = laplacian.var()

            variances.append(variance)
            
            # Create the fast fourier transform
            f = np.fft.fft2(gray)
            fshift = np.fft.fftshift(f)
            magnitude_spectrum = 20 * np.log(np.abs(fshift))

            # Calculate the mean of the magnitude spectrum
            mean = np.mean(magnitude_spectrum)

            ff_means.append(mean)

    files = np.array(files)
    variances = np.array(variances)
    ff_means = np.array(ff_means)

    df = pd.DataFrame({'Files': files, 'Variances': variances, 'FFMeans': ff_means})
    return df
```

```python
def check_blurry2(directory, file_num):
    file_num = str(file_num)

    files = []
    variances = []
    ff_means = []

    # Preallocate memory for arrays
    max_files = 200  # Adjust this number based on the expected maximum number of files
    files_arr = np.empty(max_files, dtype=object)
    variances_arr = np.empty(max_files)
    ff_means_arr = np.empty(max_files)

    # Loop through each image in the directory
    file_iter = (file_name for file_name in os.listdir(directory + file_num) if file_name.endswith((".jpg", ".png")))
    for i, file_name in enumerate(file_iter):
        files.append(file_name)

        # Load the image and convert it to grayscale
        image = cv2.imread(os.path.join(directory, file_num, file_name))
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Calculate the variance of Laplacian
        laplacian = cv2.Laplacian(gray, cv2.CV_64F)
        variance = laplacian.var()

        variances.append(variance)
        
        # Create the fast fourier transform
        f = np.fft.fft2(gray)
        fshift = np.fft.fftshift(f)
        magnitude_spectrum = 20 * np.log(np.abs(fshift))

        # Calculate the mean of the magnitude spectrum
        mean = np.mean(magnitude_spectrum)

        ff_means.append(mean)

        # Store results in preallocated arrays
        files_arr[i] = file_name
        variances_arr[i] = variance
        ff_means_arr[i] = mean

    # Create DataFrame using preallocated arrays
    df = pd.DataFrame({'Files': files_arr[:i+1], 'Variances': variances_arr[:i+1], 'FFMeans': ff_means_arr[:i+1]})
    return df
```

```python
highest_accuracy = 0.0
highest_precision = 0.0
highest_recall = 0.0
highest_f1 = 0.0
best_threshold_acc = None
best_threshold_pre = None
best_threshold_rec = None
best_threshold_f1 = None

# Iterate over the threshold values and calculate accuracy
for threshold in range(15, 250):
    # Create the 'Test' column based on the threshold condition
    full['Test'] = full['Variances'] < threshold

    # Calculate the confusion matrix
    confusion_matrix = pd.crosstab(full['Test'], full['Poor Quality'])

    # Extract the True Positives, True Negatives, and Total from the confusion matrix
    true_positives = confusion_matrix.loc[True, True]
    true_negatives = confusion_matrix.loc[False, False]
    false_positives = confusion_matrix.loc[False, True]
    false_negatives = confusion_matrix.loc[True, False]
    total = confusion_matrix.sum().sum()

    # Calculate the accuracy
    accuracy = (true_positives + true_negatives) / total
    precision = true_positives / (true_positives + false_positives)
    recall = true_positives / (true_positives + false_negatives)
    f1 = 2 * (precision * recall) / (precision + recall)

    # Check if the current accuracy is higher than the previous highest accuracy
    if accuracy > highest_accuracy:
        highest_accuracy = accuracy
        best_threshold_acc = threshold
    
    if precision > highest_precision:
        highest_precision = precision
        best_threshold_pre = threshold
        
    if recall > highest_recall:
        highest_recall = recall
        best_threshold_rec = threshold
        
    if f1 > highest_f1:
        highest_f1 = f1
        best_threshold_f1 = threshold
    

print(f"Highest Accuracy: {highest_accuracy:.2f}")
print(f"Best Threshold for Accuracy: {best_threshold_acc}")
print(f"Highest Precision: {highest_precision:.2f}")
print(f"Best Threshold for Precision: {best_threshold_pre}")
print(f"Highest Recall: {highest_recall:.2f}")
print(f"Best Threshold for Recall: {best_threshold_rec}")
print(f"Highest f1: {highest_f1:.2f}")
print(f"Best Threshold for f1: {best_threshold_f1}")
```
