```python
import os
import shutil
from PIL import Image
import pandas as pd
```

```python
# Load the metadata CSV file into a DataFrame
metadata_df = pd.read_csv('Labeling Checklist.csv')
```

```python
metadata_df
```

```python
# Drop the first 2 records
metadata_df = metadata_df.drop([0, 1]).reset_index(drop=True)
```

```python
# Filter the DataFrame to keep only records with poor_quality as true
filtered_df = metadata_df[metadata_df['POOR QUALITY'] == 'TRUE']
```

```python
filtered_df
```

```python
# Create a new folder to store the poor quality images
output_folder = 'Poor Quality Images'
os.makedirs(output_folder, exist_ok=True)
```

```python
# Iterate over each row in the DataFrame
for index, row in filtered_df.iterrows():
    folder_name = row['Folder']
    file_name = row['File']    
    # Construct the file paths
    source_path = os.path.join('Photos_all', folder_name, file_name)
    destination_folder = os.path.join(output_folder, folder_name)
    # Create the destination folder if it doesn't exist
    os.makedirs(destination_folder, exist_ok=True)
    # Copy the image file to the destination folder
    destination_path = os.path.join(destination_folder, file_name)
    shutil.copy2(source_path, destination_path)
```

```python
poor_quality_folder = 'Poor Quality Images'

# Initialize a counter
file_count = 0

# Iterate over the folders within the poor quality folder
for root, dirs, files in os.walk(poor_quality_folder):
    # Increment the file count by the number of files in each folder
    file_count += len(files)

print(f"Total number of files in all folders within {poor_quality_folder}: {file_count}")
```

```python
import os
import cv2
import numpy as np

poor_quality_folder = 'Test Images'

# Iterate over the folders within the poor quality folder
for root, dirs, files in os.walk(poor_quality_folder):
    for file in files:
        # Construct the file path
        file_path = os.path.join(root, file)

        # Read the image using OpenCV
        image = cv2.imread(file_path)

        # Apply histogram equalization to each channel
        equalized_image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
        equalized_image[:, :, 0] = cv2.equalizeHist(equalized_image[:, :, 0])
        equalized_image = cv2.cvtColor(equalized_image, cv2.COLOR_LAB2BGR)

        # Increase sharpness using Unsharp Masking
        blurred_image = cv2.GaussianBlur(equalized_image, (5, 5), 10.0)
        sharpened_image = cv2.addWeighted(equalized_image, 1.5, blurred_image, -0.5, 0)

        # Save the corrected and sharpened image back to the same file
        cv2.imwrite(file_path, sharpened_image)
```
